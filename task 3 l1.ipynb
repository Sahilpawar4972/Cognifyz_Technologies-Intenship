```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("shree.csv")
```

```python
df
```

```python

price_ranges = df['Price range'].value_counts().sort_index()
```

```python
price_ranges
```

```python
total_restaurants = len(df)
price_range_percentages = (price_ranges / total_restaurants) * 100
```

```python
price_range_percentages
```

```python
plt.figure(figsize=(10, 6))
price_ranges.plot(kind='bar')
plt.title('Distribution of Restaurant Price Ranges')
plt.xlabel('Price Range')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
bars = plt.bar(price_ranges.index, price_ranges.values)
```

```python
plt.figure(figsize=(10, 6))
bars = plt.bar(price_ranges.index, price_ranges.values)

for bar, percentage in zip(bars, price_range_percentages):
    yval = bar.get_height()
     plt.text(bar.get_x() + bar.get_width()/2, yval + 1, f'{percentage:.2f}%', ha='center', va='bottom')
plt.title('Distribution of Restaurant Price Ranges with Percentages')
plt.xlabel('Price Range')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=5)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
bars = plt.bar(price_ranges.index, price_ranges.values)

for bar, percentage in zip(bars, price_range_percentages):
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1, f'{percentage:.2f}%', ha='center', va='bottom')

plt.title('Distribution of Restaurant Price Ranges with Percentages')
plt.xlabel('Price Range')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=0)
plt.savefig('price_range_distribution.png')
plt.show()
```
