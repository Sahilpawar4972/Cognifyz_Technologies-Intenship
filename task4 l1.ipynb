```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
df=pd.read_csv("shree.csv")
```

```python
df
```

```python
total_restaurants = len(df)
```

```python
total_restaurants
```

```python
online_delivery_count = df['Has Online delivery'].str.lower().value_counts()['yes']
```

```python
online_delivery_count
```

```python
Online_delivery_percentage=(online_delivery_count/ total_restaurants) * 100
```

```python
print(f"Percentage of restaurants offering online delivery: {online_delivery_percentage:.2f}%")
```

```python
delivery_yes = df[df['Has Online delivery'].str.lower() == 'yes']
delivery_no = df[df['Has Online delivery'].str.lower() == 'no']
```

```python
average_rating_with_delivery = delivery_yes['Aggregate rating'].mean()
average_rating_without_delivery = delivery_no['Aggregate rating'].mean()
```

```python
print(f"Average rating of restaurants with online delivery: {average_rating_with_delivery:.2f}")
print(f"Average rating of restaurants without online delivery: {average_rating_without_delivery:.2f}")
```

```python
categories = ['With Delivery', 'Without Delivery']
average_ratings = [average_rating_with_delivery, average_rating_without_delivery]
```

```python
plt.figure(figsize=(8, 6))
plt.bar(categories, average_ratings, color=['green', 'red'])
plt.title('Average Ratings of Restaurants With and Without Online Delivery')
plt.xlabel('Delivery Option')
plt.ylabel('Average Rating')
plt.ylim(0, 5)
plt.show()
```
